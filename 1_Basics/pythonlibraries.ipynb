# Introduction to NumPy

Note: an array is a special variable, which can hold more than one value at a time. An array is essentially a list but holds only one data type and can perform mathematical operations.

For documentation, visit: https://numpy.org/

```python
numpy_list = [random.randint(50000,150000) for x in range(1000)]
len(numpy_list)
```

So far, we have been using these methods on one-off calculations (e.g., unique values of job_title_short, and the minimum salary).

But how can we carry out calculations on a larger scale? For instance, how can we find the minimum salary for each job type in the job_title_short column?

This is where we can use the **df.groupby()** method. The following syntax is:

- df.groupby('columns to group')['columns to aggregate'].agg_method()

 where the key parameter 'by' is used to determine the groups for the groupby.


```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

In the example above, we chose job_title_short for grouping. Therefore it will display the unique values in this column.Next we selected 'salary_year_avg' as the column to aggregate, followed by the .median() method. This finds the median values for each distinct job title and displays it.

Therefore, the aggregation column is where the calculation takes place and the result is assigned to the unique values of the groupings column.

Let's say we want to also group by country (job_country). We use the following syntax:

- df.groupby(['list of columns to group'])[['list of columns to aggregate']].agg_method()

```python
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

```python
len(df['job_country'].unique())
```

Here, job_title_short serves as the parent group. Then within a given job title such as Business Analyst, it looks through all the countries that are offering Business Analyst jobs and calculates the median value based on the country.

In other words, given that there are Business Analyst jobs in Argentina, what is the median value of these job salaries if these two conditions are True? Visually, this is like an ancestry/phylogenetic tree.

Now we know how to groupby multiple columns, but let's say we want to perform aggregations on multiple columns.

For instance, what if we want to look at (aggregate) yearly salary data and hourly salary data? We can type the following:

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

If we want to do multiple operations within the df.groupby() method, at the end rather than .median(), we can add the **.agg() method**. Syntax is as follows:

- df.groupby(['list of columns to group'])[['list of columns to aggregate']].agg(['list of aggregation methods'])

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].agg(['min', 'mean', 'median'])
```

# Matplotlib Labeling

If we remember the vertical bar plot we created, there are a lot of problems with this. Most notably, the labels in the x-axis were overlapping each other, the axis are unlabelled and there is no title.

In this section, we will learn how to clean up a visualisation.

to add a title, we have the:

- **plt.title('label')**
- **plt.ylabel('label')**

```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Number of Job Postings')
plt.xlabel('Job Type')
plt.show()
```

Now, in order to sort the overlapping job titles in the bottom, we can add format the x-axis labels 45 degrees (here, he asked CHATGPT for help with this).

It returns that we can type:
- plt.xticks(rotation = 45, ha = 'right')

```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Number of Job Postings')
plt.xlabel('Job Type')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df['job_month'] = df['job_posted_date'].dt.month
```

```python
df
```

Now, we then drop any NA values in the salary year average column and sort the values in the dataframe by job_month


```python
trend_df = df[['job_month', 'salary_year_avg']].dropna(subset='salary_year_avg')
trend_df = trend_df.sort_values(by = 'job_month')
```

```python
trend_df
```

Finally, we need to calculate the average salary per month, so we need to aggregate this data to find the mean. Here, we use the groupby function.

Note: remember to update the trend_df dataframe with the new updates before plotting

```python
trend_df = trend_df.groupby('job_month')['salary_year_avg'].mean()
```

```python
trend_df
```

```python
trend_df.max()
```

```python
trend_df.plot(x= 'job_month', y = 'salary_year_avg', kind = 'line')
plt.title('Average Job Salary Trend for 2023')
plt.ylabel('Average Monthly Salary')
plt.xlabel('Month')
plt.show()
```
