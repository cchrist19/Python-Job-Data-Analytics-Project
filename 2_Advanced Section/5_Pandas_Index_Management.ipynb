## Three Key Method To Cover

- df.reset_index(): resets the index so it goes up in iterations of one
  
- df.set_index(): sets the dataframe index using existing columns. So we can take an existing column and move it to the index.
  
- df.sort_index(): sorts values in the index

In this example, let's say we want to filter jobs that are only in the US:

```python
jobs_us = df[df['job_country'] == 'United States']
jobs_us
```

Note that because we have extracted data, the indexes do not go up in iterations of one by one (due to missing values). As a result, we can use:

- df.reset_index()

which takes many parameters, but we will focus on the inplace parameter. It asks whether to modify the dataframe rather than creating a new one and, by default, it is set to False.


```python
jobs_us.reset_index(inplace= True)
jobs_us

# Note that it kept the job_index, but has a new numerical index. We could remove 
# the job_index, but we might want to keep it if we were to merge dataframes. 
# Setting inplace to True ensures we modify the dataframe we currently have with the changes
# without having to rename it
```

The next method to look at is:

- df.set_index()

This method sets the dataframe index using existing columns. So we can take an existing column and move it to the index. It has many parameters including:

 - keys: label/column to set as the index as a string
 - inplace: This asks whether to modify the dataframe rather than creating a new one and by default is set to False. 

For instance, if we now want to move the 'job_index' column back to an index:

```python
jobs_us.set_index('job_index', inplace=True)
jobs_us 
```

 Note: by setting this column back to the index, this has displaced the unlabelled ordered indexes. So be careful with this as we lose data with this.

Now the last method is:

- df.sort_index()

which returns a new dataframe sorted by a label. Here there isn't much use case as indexes are already in ascending order.

However, in the previous example where we had pivot tables and median salaries, we can use this the sort.index() which would sort the job_salaries in alphabetical order. This can also be used when the index does not have a column title.
