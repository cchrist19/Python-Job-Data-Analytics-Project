- df.sample() method

The next method to cover is:

- df.copy(deep=True) 

which makes a copy of this objects, indices and data. When deep=True (default) a new object will be created with a copy of the calling object's data and indices (changes in the copy will not be reflected in the original)

When deep=False, a new object is created without copying the calling object's data or index (only references to the data and index are copied). Any changes to the data of the original will be reflected in the copy and vice cersa

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
# We use the .copy() method so that the original df does not get affected by updated
# changes. For instance, remember previously when we created a new df (df_filled) where 
# we replaced NA values with median values. If we were to print the original df, it 
# would reflect these changes as well. 

# This is because we set df = df_filled, so these dataframes share the same unique ID.
# This can be checked using the id(df) function.
```

We can create a new copy using the following code:

```python
df_altered = df_original.copy()
```

```python
print('ID for df_original:', id(df_original))
print('ID for df_altered:', id(df_altered))
print('Are the two dataframes the same?', id(df_original) == id(df_altered))
```

```python
print(f'ID for df_original: {id(df_original)}')
print(f'Are they the same: {id(df_original) == id(df_altered)}')

# Another way to print the df_original id using f string formatting
```
